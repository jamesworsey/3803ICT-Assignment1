```python
from __future__ import print_function, division
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk
nltk.download('punkt')
import string
from nltk.corpus import stopwords
import math
from collections import Counter
nltk.download('stopwords')
import pprint 
pp = pprint.PrettyPrinter(indent=4)

from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
def normalize(document):
    #Remove punctuation
    text = document.translate(str.maketrans('', '', string.punctuation))
    text = (text[1:])
    
    #Tokenize text
    tokens = word_tokenize(text)
    
    #Stem text
    stemmer = PorterStemmer()
    
    stem_tokens=[]
    for token in tokens:
        stem_token = stemmer.stem(token)
        stem_tokens.append(stem_token)
    ret = ' '.join(stem_tokens)
    return ret


skills = [x.strip() for x in dfICT['Requirement']]

normalized_skills = [normalize(d).split() for d in skills]
```

```python
# Flatten all the documents
flat_list = [word for skill in normalized_skills for word in skill]

# TODO: remove stop words from the vocabulary
words = [word for word in flat_list if not word in stopwords.words('english')]

# TODO: we take the 500 most common words only
counts = Counter(words)
vocabulary = counts.most_common(500)
print([x for x in vocabulary if x[0] == 'tesla'])
vocabulary = [x[0] for x in vocabulary]
assert len(vocabulary) == 500

vocabulary.sort()
vocabulary[:5]
```

```python
def tf(vocabulary, documents):
    matrix = [0] * len(documents)
    for i, document in enumerate(documents):
        counts = Counter(document)
        matrix[i] = [0] * len(vocabulary)
        for j, term in enumerate(vocabulary):
            matrix[i][j] = counts[term]
    return matrix

tf = tf(vocabulary, normalized_skills)
np.array(vocabulary)[np.where(np.array(tf[1]) > 0)], np.array(tf[1])[np.where(np.array(tf[1]) > 0)]
```

```python
def idf(vocabulary, documents):
    """TODO: compute IDF, storing values in a dictionary"""
    idf = {}
    
    #Find all documents containing word
    for i in range(len(documents)):
        for j in range(len(vocabulary)):
            if(vocabulary[j] in documents[i]):
                try:
                    idf[vocabulary[j]].add(i)
                except:
                    idf[vocabulary[j]] = {i}
    
    #Count total documents word appears in (df) then calculate idf
    for i in idf:
        N = len(documents)
        df = len(idf[i])
        idf[i] = np.log(N/(df+1))
     
    return(idf)

idf = idf(vocabulary, normalized_skills)
[idf[key] for key in vocabulary[:5]]
```

```python
def vectorize(document, vocabulary, idf):
    vector = [0]*len(vocabulary)
    counts = Counter(document)
    for i,term in enumerate(vocabulary):
        vector[i] = idf[term] * counts[term]
    return vector

document_vectors = [vectorize(s, vocabulary, idf) for s in normalized_skills]
np.array(vocabulary)[np.where(np.array(document_vectors[1]) > 1)], np.array(document_vectors[1])[np.where(np.array(document_vectors[1]) > 1)]
```
